## Tables should contain all primary, foreign and unique keys. Provide at least default values for two or more attributes in each table. Provide at least two lookup tables or check constraints where attributes have defined domains. (15 Marks)

```python
%%sql
DROP TABLE IF EXISTS Candidate_master Cascade;

CREATE TABLE Candidate_master (
CAND_ID    varchar(9)  NOT NULL,
CAND_NAME  varchar(200),
CAND_PTY_AFFILIATION  varchar(3),
CAND_ELECTION_YR  numeric(4)   DEFAULT '2008',
CAND_OFFICE_ST    varchar(2),
CAND_OFFICE       varchar(1) check(CAND_OFFICE in ('H','P','S')),
CAND_OFFICE_DISTRICT  varchar(2),
CAND_ICI      varchar(1),
CAND_STATUS   varchar(1) check(CAND_STATUS in ('F','N','P','C')),
CAND_PCC      varchar(9),
CAND_ST1      varchar(34),
CAND_ST2      varchar(34),
CAND_CITY     varchar(30),
CAND_ST       varchar(2),
CAND_ZIP      varchar(9)    DEFAULT '00000',
PRIMARY KEY (CAND_ID),
UNIQUE (CAND_ID)
);
```

```python
%%sql
DROP TABLE IF EXISTS Linkage Cascade;

CREATE TABLE Linkage (
CAND_ID varchar(9) NOT NULL,
CAND_ELECTION_YR Numeric(4) NOT NULL,
FEC_ELECTION_YR Numeric(4) NOT NULL,
CMTE_ID varchar(9),
CMTE_TP varchar(1),
CMTE_DSGN varchar(1),
LINKAGE_ID Numeric(12) NOT NULL,
PRIMARY KEY (LINKAGE_ID)
);
```

### Create a mimicking linkage table that is based on columns in both the candidate master and committee master tables

```python
%%sql
DROP TABLE IF EXISTS CACO_link Cascade;
CREATE TABLE CACO_link 
(CAND_ID,CAND_ELECTION_YR,CMTE_ID,CMTE_TP,CMTE_DSGN) AS
SELECT CA.CAND_ID, CA.CAND_ELECTION_YR, CO.CMTE_ID, CO.CMTE_TP, CO.CMTE_DSGN
FROM Candidate_master CA JOIN Committee_master CO 
on CA.CAND_ID = CO.CAND_ID;
```

### Add columns for FEC Election Year and linkage id, along with a sequence starting with one to act as the linkage id. The fec election year also is set to 2008

```python
%%sql
ALTER TABLE CACO_link 
ADD COLUMN FEC_ELECTION_YR NUMERIC,
ADD COLUMN LINKAGE_ID NUMERIC;
```

```python
%%sql
CREATE SEQUENCE seqlink
START WITH 1 
INCREMENT BY 1; 
```

```python
%%sql
UPDATE CACO_link 
SET FEC_ELECTION_YR = 2008,
LINKAGE_ID = nextval('seqlink');
```

## 5. In one query and based on the candidate’s total receipts, find the top 10 candidates from both Democratic and Republican parties. Display last name, first name, candidate state, total receipts, and total disbursement. Your answer should contain top 10 republican candidates and then top 10 democratic candidates (total of 20 rows) (10 Marks)


####  We made one query and based on the candidate’s total receipts, that finds the top 10 candidates from both Democratic and Republican parties. We displayed their last name, first name, the candidate state, the total receipts, and the total disbursement using UNION and listed in desceding order

```python
%%sql
SELECT *
FROM
    (SELECT CAND_PTY_AFFILIATION, first_name, last_name,CAND_OFFICE_ST, TTL_RECEIPTS, TTL_DISB 
     FROM All_candidate
     WHERE CAND_PTY_AFFILIATION = 'DEM' 
     ORDER BY TTL_RECEIPTS DESC
    LIMIT 10) AS D
UNION 
    (SELECT CAND_PTY_AFFILIATION, first_name, last_name,CAND_OFFICE_ST, TTL_RECEIPTS, TTL_DISB 
     FROM All_candidate
     WHERE CAND_PTY_AFFILIATION = 'REP' 
     ORDER BY TTL_RECEIPTS DESC
     LIMIT 10)
ORDER BY cand_pty_affiliation, ttl_receipts DESC;
```

## 6. In one query and based on total number of candidates, find the top 10 committee names, their committee type, total number of candidates, total number of NULL candidates, committee party affiliation (10 Marks)


#### In one query we found the top 10 committee names, their committee type, total number of candidates, total number of NULL candidates, and committee party affiliation

```python
%%sql
SELECT c.CMTE_NM, c.CMTE_TP, c.CMTE_PTY_AFFILIATION, c.cmte_id, COUNT(l.cand_id) AS total_candidate,
(count(*)-count(l.cand_id)) AS number_null from Committee_master c
RIGHT Join Linkage l on c.cmte_id = l.cmte_id
GROUP BY c.cmte_id, CMTE_NM, l.CMTE_id
ORDER BY COUNT(l.CAND_ID) desc
limit 10;
```

## 7. In on query based on contribution by Individuals, find top 10 House candidate Name, rank, State, Party, total receipts. Repeat for 10 senate candidates (10 Marks)


#### In one query we found the top 10 house candidates along with their name, rank, State, party, and total reciepts. In on query based on contribution by Individuals, find top 10 House candidate Name, rank, State, Party, total receipts. Repeat for 10 senate candidates

#### 'H' = House, 'S' = Senate

```python
%%sql
SELECT *
FROM
(SELECT CAND_OFFICE, CAND_NAME, CAND_ST, CAND_PTY_AFFILIATION, TTL_RECEIPTS, TTL_INDIV_CONTRIB
FROM
(
(SELECT Candidate_master.CAND_NAME, Candidate_master.CAND_ST, Candidate_master.CAND_PTY_AFFILIATION, 
 All_candidate.TTL_RECEIPTS, Candidate_master.CAND_OFFICE, All_candidate.TTL_INDIV_CONTRIB
FROM All_candidate, Candidate_master
WHERE All_candidate.CAND_ID = Candidate_master.CAND_ID)
) C
WHERE C.CAND_OFFICE = 'H'
ORDER BY C.TTL_INDIV_CONTRIB DESC
LIMIT 10) AS H
UNION
(SELECT CAND_OFFICE, CAND_NAME, CAND_ST, CAND_PTY_AFFILIATION, TTL_RECEIPTS, TTL_INDIV_CONTRIB
FROM
(
(SELECT Candidate_master.CAND_NAME, Candidate_master.CAND_ST, Candidate_master.CAND_PTY_AFFILIATION, 
 All_candidate.TTL_RECEIPTS, Candidate_master.CAND_OFFICE, All_candidate.TTL_INDIV_CONTRIB
FROM All_candidate, Candidate_master
WHERE All_candidate.CAND_ID = Candidate_master.CAND_ID)
) C
WHERE C.CAND_OFFICE = 'S'
ORDER BY C.TTL_INDIV_CONTRIB DESC
LIMIT 10)
ORDER BY CAND_OFFICE, TTL_INDIV_CONTRIB DESC
;
```

## 8. Write a view that provide top 10 Republican committees in terms of the total number of individuals for those committees. Do the same for democratic committees. Calculate the total number of transaction receipts by those top 20 committees and sort the final answer based on total receipts. (10 marks)


#### We made a view that provides the top 10 Republican committees in terms of the total number of individuals for those committees as well as a similar view for the top 10 Democratic committees 

##### Top 10 Republican committees in terms of the total number of individual for those committees.

```python
%%sql
DROP VIEW IF EXISTS TOP10_Republican;
CREATE OR REPLACE VIEW TOP10_Republican AS
SELECT IR.id, count(distinct IR.name) AS ttl_num_ind
FROM (
    select Con_by_ind.cmte_id AS id , Con_by_ind.name AS name
    FROM Con_by_ind
    where Con_by_ind.cmte_id IN(
        select Committee_master.cmte_id
        FROM Committee_master 
        where Committee_master.CMTE_PTY_AFFILIATION='REP')
) AS IR
GROUP BY IR.id
ORDER BY ttl_num_ind DESC
LIMIT 10;
```

##### Top 10 Democratic committees in terms of the total number of individual for those committees.

```python
%%sql
DROP VIEW IF EXISTS TOP10_Democratic;
CREATE OR REPLACE VIEW TOP10_Democratic AS
select ID.id, count(distinct ID.name) as ttl_num_ind
from (
    select Con_by_ind.cmte_id as id , Con_by_ind.name as name
    from Con_by_ind
    where Con_by_ind.cmte_id in(
        select Committee_master.cmte_id
        from Committee_master 
        where Committee_master.CMTE_PTY_AFFILIATION='DEM')
) as ID
group by ID.id
order by ttl_num_ind DESC
limit 10;
```

#### We then found total number of transaction receipts by those top 20 committees and sorted the final answer based on total receipts

```python
%%sql
SELECT RD1.cmte_id, Committee_master.cmte_nm,Committee_master.cmte_pty_affiliation,RD1.number_of_receipts, RD1.ttl_receipts
FROM
(
SELECT RD.cmte_id, count(RD.*) as number_of_receipts, sum(RD.transaction_amt) as ttl_receipts
FROM
    (
    SELECT * 
    FROM Con_by_ind
    WHERE cmte_id IN (
        SELECT TOP10_Republican.id 
        FROM TOP10_Republican
        UNION
        SELECT TOP10_Democratic.id
        FROM TOP10_Democratic
    )
) as RD 
GROUP BY RD.cmte_id
ORDER BY ttl_receipts DESC) as RD1, Committee_master
WHERE RD1.cmte_id = Committee_master.cmte_id;
```
